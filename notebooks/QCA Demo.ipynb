# QCA
This notebook demonstrates calling the QCA REST endpoint.

```python
import requests
import json
```

# Call the REST method
*Note: The Rest endpoint is running on a development server and is subject to down time and removal at any point*


* ## URL   
    http://ec2-52-37-182-174.us-west-2.compute.amazonaws.com:5602/directedpath/query

* ## Method    
    POST
    
* ## URL Params     
    #### Required:    
        source=[whitespace separated list of nodes]
        target=[whitespace separated list of nodes]
        networkcx=[cx file]
        
    #### Optional:
        pathnum=[integer]

* ## Success Response:
        Code: 200
        Content: JSON

* ## Error Response:
        Code: 400 (Bad request)
        Content: Error message contained in JSON
        

```python
base_url = 'http://ec2-52-37-182-174.us-west-2.compute.amazonaws.com:5602/directedpath/query'

source = 'EGFR'
target = 'MAP2K1 MAP2K2'
max_number_of_paths = 5

url = base_url + '?source=' + source + '&target=' + target + '&pathnum=' + str(max_number_of_paths)

files = {'networkcx': open('The_RAS_Machine.cx', 'rb')}

r = requests.post(url, files=files)
```

## Output the results

```python
d = json.loads(r.text)
print json.dumps(d, indent=4)
```
